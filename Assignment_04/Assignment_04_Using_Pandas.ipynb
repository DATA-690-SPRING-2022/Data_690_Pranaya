## 8.Clean the data to make the year, population, cost, cost per person numerical type - use apply() function


```python
df["cleaned_year"]=df["Census Year"].apply(lambda x : x.replace("*",""))
df["cleaned_year"]
```

```python
df["cleasned_pop"]=df["Total Population"].apply(lambda x:x.replace(",",""))
df["cleasned_pop"]
```

```python
df["cleasned_cost"]=df["Census Cost"].apply(lambda x:x.replace(",","").replace("$","").replace(" ",""))
df["cleasned_cost"]
```

```python
df["cleasned_cost"]=df["cleasned_cost"].apply(lambda x:x.replace("Billion","000000000000").replace(".50","5"))
df["cleasned_cost"]
```

```python
df["cleansed_avg"] = df["Average Cost Per Person"].apply(lambda x:x.replace('$','').replace('cents',''))
df['cleansed_avg']
```

## 9.Add the clean data into new Data Frame

```python
df2=df[df.columns[4:]]
df2
```

## 10. Change the Data type to Int

```python
df2["cleasned_cost"]=df2["cleasned_cost"].astype(int)
df2["cleaned_year"]=df2["cleaned_year"].astype(int)
df2["cleasned_pop"]=df2["cleasned_pop"].astype(int)
df2["cleansed_avg"]=df2["cleansed_avg"].astype(float)
df2
```
